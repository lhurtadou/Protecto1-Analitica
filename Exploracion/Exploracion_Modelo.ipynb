# Análisis del Histograma de las Variables Numéricas

Este histograma presenta la distribución de varias variables numéricas en el conjunto de datos. A continuación, se proporciona una interpretación para cada una de las variables mostradas:

## 1. **Rented Bike Count (Número de bicicletas rentadas)**
- **Distribución**: Tiene una distribución asimétrica hacia la derecha, lo que indica que en la mayoría de los casos se rentan pocas bicicletas (entre 0 y 1000), mientras que es raro ver números de alquileres muy altos (más de 2000).
- **Interpretación**: La mayoría de las personas probablemente alquilan pocas bicicletas en la mayoría de las horas o días del año.

## 2. **Hour (Hora)**
- **Distribución**: La distribución de las horas es bastante uniforme, lo que indica que hay un número más o menos constante de registros para cada hora del día.
- **Interpretación**: Se están capturando datos de alquileres durante todo el día, sin ninguna hora especialmente predominante en cuanto a la frecuencia de alquileres.

## 3. **Temperature(C) (Temperatura en grados Celsius)**
- **Distribución**: Tiene una distribución cercana a la normal (campana), con un pico alrededor de los 15-20°C.
- **Interpretación**: La temperatura común en este conjunto de datos suele estar entre los 10°C y los 20°C, con muy pocos registros en temperaturas extremadamente bajas o altas.

## 4. **Humidity(%) (Humedad en porcentaje)**
- **Distribución**: La humedad muestra una distribución algo uniforme, aunque ligeramente sesgada hacia la derecha. Hay más registros con niveles de humedad entre el 40% y el 80%.
- **Interpretación**: Esto sugiere que el clima en general tiene un rango de humedad moderado, con menos ocurrencias de niveles muy bajos o muy altos de humedad.

## 5. **Wind speed (m/s) (Velocidad del viento en metros por segundo)**
- **Distribución**: Asimétrica hacia la derecha. La mayoría de los registros tienen velocidades del viento bajas (entre 0 y 2 m/s).
- **Interpretación**: La velocidad del viento es baja la mayor parte del tiempo, y no se registran muchas velocidades del viento superiores a los 4 m/s.

## 6. **Visibility (10m) (Visibilidad en decenas de metros)**
- **Distribución**: Muy sesgada hacia la derecha. La mayoría de los registros tienen una visibilidad de 2000 (máximo), y pocas veces la visibilidad es menor.
- **Interpretación**: La visibilidad es muy buena (2000 decenas de metros, equivalente a 20 kilómetros) en la mayoría de los casos, lo que podría ser indicativo de condiciones atmosféricas estables.

## 7. **Dew point temperature(C) (Temperatura del punto de rocío en grados Celsius)**
- **Distribución**: Tiene una distribución casi uniforme, aunque con un pico notable entre los 0°C y los 5°C.
- **Interpretación**: La mayoría de las temperaturas de punto de rocío están por encima de los -10°C, lo que sugiere un clima donde la humedad no suele condensarse a temperaturas muy bajas.

## 8. **Solar Radiation (MJ/m²) (Radiación solar en MJ/m²)**
- **Distribución**: La mayoría de los valores están concentrados cerca de 0, lo que indica que la radiación solar suele ser baja la mayor parte del tiempo.
- **Interpretación**: Es posible que muchos de los registros se hayan tomado en días nublados o en épocas del año con poca luz solar, o tal vez durante la noche.

## 9. **Rainfall (mm) (Precipitación en milímetros)**
- **Distribución**: Tiene una distribución extremadamente sesgada a la derecha. La mayoría de los registros muestran 0 mm de precipitación.
- **Interpretación**: La mayoría de los días no llueve o llueve muy poco, con unos pocos días en los que hay precipitaciones mayores a 10 mm.

## 10. **Snowfall (cm) (Nevadas en centímetros)**
- **Distribución**: También muy sesgada hacia la derecha, con la mayoría de los días mostrando 0 cm de nieve, y solo unos pocos días con nevadas significativas.
- **Interpretación**: La mayoría de los días no hay nevadas, y cuando las hay, son mínimas. Sin embargo, en casos raros, hay acumulaciones notables de nieve (más de 5 cm).


```python
# Diagramas de caja para identificar valores atípicos
plt.figure(figsize=(15, 10))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']),palette='Set1')
plt.title('Diagrama de Caja de las Variables Numéricas', fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

# Análisis del Diagrama de Caja de las Variables Numéricas

Este diagrama de caja (boxplot) muestra la dispersión de varias variables numéricas en el conjunto de datos, resaltando sus valores atípicos, medianas y rangos intercuartiles.

## 1. **Rented Bike Count (Número de bicicletas rentadas)**
- **Descripción**: La caja muestra una gran dispersión en el número de bicicletas rentadas. Los valores van desde aproximadamente 0 hasta más de 1000, y hay muchos valores atípicos (outliers) por encima de los 2500.
- **Interpretación**: Esto sugiere que, aunque la mayoría de los registros tienen entre 0 y 1000 bicicletas rentadas, existen bastantes registros con un número excepcionalmente alto de bicicletas alquiladas, que son considerados valores atípicos.

## 2. **Hour (Hora)**
- **Descripción**: La caja es muy pequeña, lo que sugiere una distribución muy uniforme con poca variabilidad en la variable.
- **Interpretación**: No hay valores atípicos en la variable "Hora", lo que indica que las horas están distribuidas uniformemente, probablemente en un rango de 0 a 23 horas.

## 3. **Temperature(C) (Temperatura en grados Celsius)**
- **Descripción**: La dispersión de la temperatura es baja, con la mayoría de los valores concentrados entre -5°C y 30°C. No hay valores atípicos visibles.
- **Interpretación**: La temperatura tiene una distribución moderada y no presenta registros extremos o atípicos en los datos.

## 4. **Humidity(%) (Humedad en porcentaje)**
- **Descripción**: La caja muestra una dispersión moderada en la variable de humedad, con valores que oscilan entre 20% y 100%. Existen algunos valores atípicos a niveles más bajos de humedad.
- **Interpretación**: Esto indica que, aunque la mayoría de los registros están dentro del rango esperado, algunos casos de baja humedad son poco comunes y aparecen como valores atípicos.

## 5. **Wind speed (m/s) (Velocidad del viento en metros por segundo)**
- **Descripción**: La dispersión es pequeña y presenta algunos valores atípicos con velocidades del viento superiores a 4 m/s.
- **Interpretación**: La mayoría de los registros tienen velocidades del viento bastante bajas, pero hay unos pocos casos donde la velocidad del viento es inusualmente alta.

## 6. **Visibility (10m) (Visibilidad en decenas de metros)**
- **Descripción**: La caja es bastante alta, lo que indica una variabilidad considerable en la visibilidad, con valores que varían entre 500 y 2000 decenas de metros.
- **Interpretación**: La visibilidad varía significativamente a lo largo de los registros, lo que sugiere que las condiciones atmosféricas afectan la visibilidad de manera notable en diferentes momentos.

## 7. **Dew point temperature(C) (Temperatura del punto de rocío en grados Celsius)**
- **Descripción**: La dispersión es baja, con un rango de valores concentrados entre -20°C y 20°C. No hay valores atípicos significativos.
- **Interpretación**: La temperatura del punto de rocío está distribuida de manera moderada, sin valores extremos.

## 8. **Solar Radiation (MJ/m²) (Radiación solar en MJ/m²)**
- **Descripción**: Hay una pequeña dispersión en los valores de radiación solar, con algunos valores atípicos por encima del rango común.
- **Interpretación**: La mayoría de los registros muestran poca radiación solar, pero existen algunos momentos con niveles notablemente altos de radiación.

## 9. **Rainfall (mm) (Precipitación en milímetros)**
- **Descripción**: La caja está prácticamente en el rango de 0 mm, lo que indica que la mayoría de los registros no presentan precipitación, pero hay algunos valores atípicos de lluvia más intensa.
- **Interpretación**: La lluvia es rara en este conjunto de datos, pero cuando ocurre, se registra en valores inusualmente altos, lo que indica un comportamiento atípico en algunos momentos.

## 10. **Snowfall (cm) (Nevadas en centímetros)**
- **Descripción**: La dispersión es casi inexistente, lo que indica que la mayoría de los registros no tienen nevadas, pero hay unos pocos valores atípicos con nevadas superiores a 5 cm.
- **Interpretación**: Las nevadas son poco comunes, pero cuando ocurren, se registran valores significativamente altos en pocos casos.

---

## **Conclusiones Generales del Diagrama de Caja**
- **Variabilidad**: La variable **Rented Bike Count** muestra una alta variabilidad, con numerosos valores atípicos hacia valores altos de alquiler de bicicletas.
- **Condiciones Climatológicas**: Las variables como la **velocidad del viento**, **visibilidad**, y **precipitación** presentan algunas variaciones interesantes con valores atípicos notables, lo que sugiere que las condiciones extremas son poco frecuentes pero presentes en el conjunto de datos.
- **Distribución Uniforme**: Algunas variables como la **hora** y **temperatura** tienen poca variabilidad, lo que indica una distribución uniforme en el caso de la hora, y una distribución moderada en el caso de la temperatura.
